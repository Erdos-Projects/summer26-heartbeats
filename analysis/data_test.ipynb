This is a short notebook showing how to extract activity interval data using the extract_intervals function in read_data.py

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from read_data import extract_intervals, get_activityMap, interval_stats
```

We can use the extract_intervals function to create a dataframe of activity intervals, labeled by subject ID and the startstop timestamps. It can take 30-60 seconds to read in everything and create the new dataframe, but we can save this as a separate file if we want to speed things up.

```python
#create a dataframe for activity intervals and populate it with data from all subjects

df_intervals = pd.DataFrame(columns=['activity_id','subject_id','length','t1','t2'])

folder_path='../data/PAMAP2_Dataset/Protocol/'
for subject_num in range(105, 109):
    df_intervals = extract_intervals(str(subject_num), df_intervals, folder_path)
```

Once we create the interval dataframe, we can extract the data as needed. Here I make some histograms of activity length for each activity. My script only reads in the 12 procedure activities, so the optional ones are empty.

```python
#map from activity IDs to string labels
activity_map = get_activityMap()

#generate histograms of activity periods
fig,axarr = plt.subplots(4,5,figsize=[24,18])

i = 0
j = 0

for activity in activity_map.keys():
    axarr[i,j].hist(df_intervals[df_intervals['activity_id']==activity]['length'],bins=5, edgecolor='black')
    axarr[i,j].set_title(activity_map[activity])
    axarr[i,j].set_xlabel('Activity Length (s)')
    j=(j+1)%5
    if j == 0:
        i+=1

plt.tight_layout()
plt.show()
```

```python

for activity_id in activity_map.keys():
    interval_stats(df_intervals,activity_id)
```
