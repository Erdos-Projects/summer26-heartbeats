This is a short notebook showing how to extract activity interval data using the extract_intervals function in read_data.py

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from read_data import extract_intervals, get_activityMap
```

```python
#create a dataframe for activity intervals and populate it with data from all subjects

df_intervals = pd.DataFrame(columns=['activity_id','subject_id','length','t1','t2'])

df_intervals = extract_intervals('101',df_intervals,verbose=False)
df_intervals = extract_intervals('102',df_intervals,verbose=False)
df_intervals = extract_intervals('103',df_intervals,verbose=False)
df_intervals = extract_intervals('104',df_intervals,verbose=False)
df_intervals = extract_intervals('105',df_intervals,verbose=False)
df_intervals = extract_intervals('106',df_intervals,verbose=False)
df_intervals = extract_intervals('107',df_intervals,verbose=False)
df_intervals = extract_intervals('108',df_intervals,verbose=False)
df_intervals = extract_intervals('109',df_intervals,verbose=False)
```

```python
#map from activity IDs to string labels
activity_map = get_activityMap()

#generate histograms of activity periods
fig,axarr = plt.subplots(4,5,figsize=[24,18])

i = 0
j = 0

for activity in activity_map.keys():
    axarr[i,j].hist(df_intervals[df_intervals['activity_id']==activity]['length'],bins=5, edgecolor='black')
    axarr[i,j].set_title(activity_map[activity])
    axarr[i,j].set_xlabel('Activity Length (s)')
    j=(j+1)%5
    if j == 0:
        i+=1

plt.tight_layout()
plt.show()
```
